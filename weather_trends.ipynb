# Moving Averages

Moving averages are used to smooth out data to make it easier to observe long term trends and not get lost in daily fluctuations. For example, let's say you wanted to visualize the sales trend at a clothing retail store. You start with daily data, and your chart looks too volatile to interpret because more people shop on the weekends, so sales spike on those days.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

Istanbul is the city where I live so i downloaded all the rows for it and saved it in a csv file to load into a DF

```python
istanbul = pd.read_csv('city_data.csv') #istanbul df
gd = pd.read_csv('global_data.csv') #global_data df 
```

### Line Plot for Istanbul without implementing moving average

```python
# line plot - the yearly average air temperature in Istanbul
istanbul.plot(x = 'year', y='avg_temp', color='green', linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend('')

# # title and labels
plt.title('The yearly average air temperature in Istanbul', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Average Temperature [°C]', fontsize=16);
```

### Line Plot for global air temperature without implementing moving average

```python
# line plot - the yearly average gloabal air temperature
gd.plot(x = 'year', y='avg_temp', color='green', linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend('')

# # title and labels
plt.title('The yearly global average air temperature', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Average Temperature [°C]', fontsize=16);
```

```python
istanbulMA = istanbul.copy()
```

```python
istanbulMA.rolling(7).mean()
```

```python
istanbul['MA'] = istanbulMA.rolling(7).mean()['avg_temp']
```

```python
istanbul
```

### Line Plot for Istanbul with implementing moving average

```python
# line plot - the yearly average air temperature in Istanbul
istanbul.plot(x = 'year', y='MA', color='green', linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend('')

# # title and labels
plt.title('The yearly average air temperature in Istanbul', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Average Temperature [°C]', fontsize=16);
```

```python
gdMA = gd.copy()
```

```python
gd['MA'] = gdMA.rolling(7).mean()['avg_temp']
```
